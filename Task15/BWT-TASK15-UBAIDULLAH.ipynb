### 1. Merge two DataFrames on a single key.


```python
#Dataset1
import pandas as pd
file_path = r'E:\IntenshipDatasets\healthcare_dataset.csv'
df1 = pd.read_csv(file_path, nrows=500)
df1.head()
```

```python
# Dataset2
file_path = r'E:\IntenshipDatasets\medical_records.csv'
df2 = pd.read_csv(file_path, nrows=500)
df2.head()
```

### Dataset2 is not prepare for merging first we prepare the dataset.
###  Capitalize the first letter of each column name and convert the rest to lowercase.

```python
df2.columns = df2.columns.str.capitalize()
df2.head()
```

###  Replace 'M' with 'Male' and 'F' with 'Female' in the 'gender' column

```python
df2['Gender'] = df2['Gender'].replace({'M': 'Male', 'F': 'Female'})
df2.head()
```

### Merge df1 and df2 on the 'Gender' column

```python
merged_df = pd.merge(df1, df2, on='Gender', suffixes=('_df1', '_df2'))
merged_df
```

### 3. Perform an outer join, inner join, left join, and right join.


```python
file_path_employ = r'E:\IntenshipDatasets\Employ.csv'
file_path_salaries = r'E:\IntenshipDatasets\Salaries.csv'
file_path_departments = r'E:\IntenshipDatasets\Department.csv'

df_employ = pd.read_csv(file_path_employ, nrows=500)
df_salaries = pd.read_csv(file_path_salaries, nrows=500)
df_departments = pd.read_csv(file_path_departments, nrows=500)

print("Employees DataFrame:")
print(df_employ.head())
print("\nSalaries DataFrame:")
print(df_salaries.head())
print("\nDepartments DataFrame:")
print(df_departments.head())
```

### Outer Join

```python
outer_join = pd.merge(df_employ, df_salaries, on='Emp_ID', how='outer')
print("\nOuter Join Result:")
print(outer_join.head())
```

### Inner Join

```python
inner_join = pd.merge(df_employ, df_salaries, on='Emp_ID', how='inner')
print("\nInner Join Result:")
print(inner_join.head())
```

### Left Join

```python
left_join = pd.merge(df_employ, df_salaries, on='Emp_ID', how='left')
print("\nLeft Join Result:")
print(left_join.head())
```

### Right Join

```python
right_join = pd.merge(df_employ, df_salaries, on='Emp_ID', how='right')
print("\nRight Join Result:")
print(right_join.head())
```

### Merging employees with departments

```python
emp_dept_merge = pd.merge(df_employ, df_departments, on='Department', how='inner')
print("\nEmployees and Departments Merge Result:")
print(emp_dept_merge.head())
```

### 7. Reshape data using the melt function to go from wide to long format.

```python
file_path_students= r'E:\IntenshipDatasets\Studentdata.csv'

df_students = pd.read_csv(file_path_students, nrows=500)

print("Student DataFrame:")
print(df_students.head())
```

```python
df_long = pd.melt(df_students, id_vars=['ID', 'Name'], 
                  var_name='Subject', value_name='Score')

print("\nReshaped DataFrame (Long Format):")
print(df_long)
```

### 8. Create a pivot table to summarize data.


```python
pivot_table = df_long.pivot_table(index=['Subject'], columns=['ID'], values='Score', aggfunc='mean')

print("\nPivot Table:")
print(pivot_table)
```
