## **Function: img2txt**

`def img2txt(input_text, input_image):` This function takes an input text and an image file path, processes the image to generate a descriptive text based on the input, and logs relevant information.

 **1. Loads the Image**
- Opens the specified image file.

 **2. Logs Input Details**
- Records the input text and its details in a log file.

 **3. Sets Prompt Instructions**
- Prepares a prompt for image description based on the type of `input_text`.

 **4. Generates Output**
- Uses a pre-defined pipeline to generate descriptive text from the image.

 **5. Extracts Response**
- Extracts and returns the assistant's response from the generated output.


```python
def img2txt(input_text, input_image):
    # Load the image
    image = Image.open(input_image)

    writehistory(f"Input text: {input_text} - Type: {type(input_text)} - Dir: {dir(input_text)}")

    if isinstance(input_text, tuple):
        prompt_instructions = """
        Describe the image using as much detail as possible, is it a painting, a photograph, what colors are predominant, what is the image about?
        """
    else:
        prompt_instructions = """
        Act as an expert in imagery descriptive analysis, using as much detail as possible from the image, respond to the following prompt:
        """ + input_text

    writehistory(f"prompt_instructions: {prompt_instructions}")
    prompt = "USER: <image>\n" + prompt_instructions + "\nASSISTANT:"

    outputs = pipe(image, prompt=prompt, generate_kwargs={"max_new_tokens": 200})

    # Properly extract the response text
    if outputs is not None and len(outputs[0]["generated_text"]) > 0:
        match = re.search(r'ASSISTANT:\s*(.*)', outputs[0]["generated_text"])
        if match:
            # Extract text after ASSISTANT:
            reply = match.group(1)
        else:
            reply = "No response found"
    else:
        reply = "No response generated"

    return reply
```

## **Function: transcribe**

`def transcribe(audio):` This function takes an audio input, processes it using the Whisper model, and returns the transcribed text.

 **1. Checks Audio Input**
- Verifies if the audio input is empty or `None`, returning empty values in that case.

 **2. Loads and Processes Audio**
- Loads the audio file, pads or trims it for uniform length, and computes the log-mel spectrogram.

**3. Detects Language**
- Uses the model to detect the language of the audio.

 **4. Decodes Audio**
- Decodes the processed audio to generate the transcription.

 **5. Returns Transcription**
- Finally, it returns the transcribed text.


```python
def transcribe(audio):

    # Check if the audio input is None or empty
    if audio is None or audio == '':
        return ('','',None)  # Return empty strings and None audio file

    # language = 'en'

    audio = whisper.load_audio(audio)
    audio = whisper.pad_or_trim(audio)

    mel = whisper.log_mel_spectrogram(audio).to(model.device)

    _, probs = model.detect_language(mel)

    options = whisper.DecodingOptions()
    result = whisper.decode(model, mel, options)
    result_text = result.text

    return result_text
```

## **Function: text_to_speech**

`def text_to_speech(text, file_path):` This function converts the provided text into speech and saves it as an audio file.

**1. Sets Language**
- Specifies the language for the speech (English in this case).

**2. Creates Speech Object**
- Uses `gTTS` to create a text-to-speech object with the given text and language settings.

**3. Saves Audio File**
- The generated speech is saved to the specified file path.

**4. Returns File Path**
- Finally, it returns the path of the saved audio file.


```python
def text_to_speech(text, file_path):
    language = 'en'

    audioobj = gTTS(text = text,
                    lang = language,
                    slow = False)

    audioobj.save(file_path)

    return file_path
```

## **Gradio Interface for Speech-to-Text and Image Processing**

This code sets up a Gradio interface that seamlessly combines speech-to-text and image processing functionalities.

**1. Imports Libraries**
- Utilizes essential libraries:
  - Gradio for user interface
  - `base64` for encoding
  - `os` for file handling

**2. Processes Inputs**
- The `process_inputs` function handles:
  - Audio and image paths
  - Transcribes audio via the `transcribe` function
  - Generates descriptive text from the image using `img2txt`
    - Returns a message if no image is provided
  - Converts output text to speech, saving it as an audio file

**3. Creates Gradio Interface**
- Configures:
  - Audio input from the microphone
  - Image file input
- Specifies outputs:
  - Text boxes for results
  - An audio player for the generated speech

**4. Launches Interface**
- The Gradio interface is launched in debug mode for testing.


```python
import gradio as gr
import base64
import os

# A function to handle audio and image inputs
def process_inputs(audio_path, image_path):
    # Process the audio file (assuming this is handled by a function called 'transcribe')
    speech_to_text_output = transcribe(audio_path)

    # Handle the image input
    if image_path:
        chatgpt_output = img2txt(speech_to_text_output, image_path)
    else:
        chatgpt_output = "No image provided."

    # Assuming 'transcribe' also returns the path to a processed audio file
    processed_audio_path = text_to_speech(chatgpt_output, "Temp3.mp3")  # Replace with actual path if different

    return speech_to_text_output, chatgpt_output, processed_audio_path

# Create the interface
iface = gr.Interface(
    fn=process_inputs,
    inputs=[
        gr.Audio(sources=["microphone"], type="filepath"),
        gr.Image(type="filepath")
    ],
    outputs=[
        gr.Textbox(label="Speech to Text"),
        gr.Textbox(label="ChatGPT Output"),
        gr.Audio("Temp.mp3")
    ],
    title="Learn OpenAI Whisper: Image processing with Whisper and Llava",
    description="Upload an image and interact via voice input and audio response."
)

# Launch the interface
iface.launch(debug=True)
```
