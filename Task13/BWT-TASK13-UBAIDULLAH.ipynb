```python
def mean_range(series):
    if pd.api.types.is_numeric_dtype(series):
        return np.mean(series.max() - series.min())
    else:
        return None  # Handle non-numeric columns gracefully

# Applying the custom aggregation function
grouped = df.groupby('job')['age'].agg(mean_range)
print("Grouped by 'job' and applied custom aggregation function (mean range) to 'age':")
grouped
```

### 9. Group the DataFrame by multiple columns and calculate the sum of each group.


```python
# Group by 'job' and 'education', calculate sum of 'age' and 'balance'
grouped = df.groupby(['job', 'education']).agg({'age': 'sum', 'balance': 'sum'})
print("Grouped by 'job' and 'education', calculated sum of 'age' and 'balance':")
grouped
```

```python
# Define a custom function for aggregation
def sum_columns(group):
    return pd.Series({
        'total_age': group['age'].sum(),
        'total_balance': group['balance'].sum()
    })

# Suppress FutureWarning and DeprecationWarning
import warnings
with warnings.catch_warnings():
    warnings.simplefilter('ignore', FutureWarning)
    warnings.simplefilter('ignore', DeprecationWarning)
    
    # Group by 'job' and 'education', apply custom aggregation function
    grouped = df.groupby(['job', 'education'], as_index=False, group_keys=False).apply(sum_columns)

# Print the result
print("Grouped by 'job' and 'education', applied custom aggregation function to sum 'age' and 'balance':")
print(grouped)
```

### 10. Use the agg method to apply multiple aggregation functions to grouped data.


```python
# Group by 'job' and 'education', apply multiple aggregation functions
grouped = df.groupby(['job', 'education']).agg({
    'age': ['mean', 'max', 'min'],
    'balance': ['sum', 'mean']
}).reset_index()

print(grouped)
```

```python
# Group by 'job' and 'education', apply multiple aggregation functions with lambda functions
grouped = df.groupby(['job', 'education']).agg({
    'age': lambda x: x.mean() if 'mean' in x.name else x.max() if 'max' in x.name else x.min() if 'min' in x.name else None,
    'balance': lambda x: x.sum() if 'sum' in x.name else x.mean() if 'mean' in x.name else None
}).reset_index()

print(grouped)
```

```python
# Custom function to calculate range
def range_func(x):
    return x.max() - x.min()

# Group by 'job' and 'education', apply multiple aggregation functions using named aggregation
grouped = df.groupby(['job', 'education']).agg(
    mean_age=('age', 'mean'),
    max_age=('age', 'max'),
    min_age=('age', 'min'),
    total_balance=('balance', 'sum'),
    range_balance=('balance', range_func)
).reset_index()

print(grouped)
```

```python
# Group by 'job' and 'education', apply different aggregation functions to different columns
grouped = df.groupby(['job', 'education']).agg({
    'age': 'mean',
    'balance': lambda x: x.max() - x.min()  # Custom aggregation function
}).reset_index()

print(grouped)
```

```python
# Group by 'job' and 'education', apply a list of aggregation functions
grouped = df.groupby(['job', 'education']).agg({
    'age': ['mean', 'median'],
    'balance': ['sum', 'std']
}).reset_index()

print(grouped)
```

```python
from functools import partial

# Partial function to calculate difference
difference_func = partial(lambda x, func: func(x), func=lambda x: x.max() - x.min())

# Group by 'job' and 'education', apply multiple aggregation functions
grouped = df.groupby(['job', 'education']).agg({
    'age': ['mean', 'max', 'min', difference_func],
    'balance': ['sum', 'mean']
}).reset_index()

print(grouped)
```

```python
# Group by 'job' and 'education', apply multiple aggregation functions
grouped = df.groupby(['job', 'education']).agg({
    'age': lambda x: x.mean() if 'mean' in x.name else x.max() if 'max' in x.name else x.min() if 'min' in x.name else None,
    'balance': lambda x: x.sum() if 'sum' in x.name else x.mean() if 'mean' in x.name else None
}).reset_index()

print(grouped)
```
